## Load IOUs

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

PROJECT_ROOT = Path(globals()['_dh'][0]).resolve().parent
DATA_PATH = PROJECT_ROOT / "data"

IOUS_PATH = DATA_PATH / "detection"  / "ious.npy"
ious = np.load(IOUS_PATH)
```

## Histogram of IOUs

```python
plt.hist(ious, bins=100)
plt.title("Histogram of IOUs")
plt.xlabel("IOU")
plt.ylabel("Count")
plt.show()
```

### Histogram of IOUs smaller than 1.0

```python
# plot histogram of ious
nonperfect_ious = ious[ious != 1.0]
plt.hist(nonperfect_ious, bins=100)
plt.title("Histogram of IOUs < 1.0")
plt.xlabel("IOU")
plt.ylabel("Count")
plt.show()
```

### Percentages of perfect, true detections and false detections

```python
perfect_ious = ious[ious == 1.0]
print(f"Number of perfect IOUs: {len(perfect_ious)}, {len(perfect_ious) / len(ious) * 100 :.3f} %")
print(f"Number of non-perfect IOUs: {len(nonperfect_ious)}, {len(nonperfect_ious) / len(ious) * 100 :.3f} %")
print(f"Number of IOUs < 0.5: {len(nonperfect_ious[nonperfect_ious < 0.5])}, {len(nonperfect_ious[nonperfect_ious < 0.5]) / len(ious) * 100 :.3f} %")
```

### Histogram of IOUs < 0.5

```python
nonperfect_ious = ious[ious < 0.5]
plt.hist(nonperfect_ious, bins=100)
plt.title("Histogram of IOUs < 0.5")
plt.xlabel("IOU")
plt.ylabel("Count")
plt.show()
```
